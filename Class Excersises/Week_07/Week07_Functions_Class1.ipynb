## Top three reasons to write a function

The top three reasons to write a function are, in no particular order:
1. To keep from writing the same code in several places. Write it once, then call it when you need to. This adheres to the coding commandment abbreviated DRY: **Don't Repeat Yourself.** If you put the code in once place, you only have to edit in one place.
2. **Modularity.** If each function does one task, then you know where to go to find all the code for that one task -- there's no other code hiding out somewhere that you forgot about.  Big codes do complicated things. Functions help you break them up into small steps that can be executed when you need them.
3. **Encapsulation** The variables you define inside a function stay in that function, per Las Vegas tourism board mandate. You don't have to worry about calling a variable with the same name you used 100 lines ago, overwriting its value.

As you start to write longer and more complex code, you'll find yourself re-using a few lines or a lot of lines, often with minor tweaks between the uses.  Whenever you think about copying and pasting, consider writing a function first!  

## Python function definition syntax

The function above is pretty simple and just says hello.  It's a version of a function with no inputs or outputs, it just does a task.  The syntax goes:

- \<def\> the keywork that defines the function
- \<say_hello\> the name of the function
- \<()\> some parentheses. We'll put the parameters in here later.
- \<colon>\> punctuation that tells Python this first line is done
- \<indent\> starts the body of the function. everything that's indented this same amount belongs to the function
- \< a blank line at the end \> optional but good style

## Control Flow 

The function must be declared before it's run, so they usually all go at the top of your script, right after the `import` statements if there are any.  Your script starts executing code line-by-lane after the function definitions, where you first encounter left-justified lines of code.  When a function is called, you skip up to the function definition and start executing lines of the function. When you're done, you return to the main body of text and continue with the next line.

## Parameters and arguments

Most of the functions that you're familiar with take an input and produce an output. Examples include a square root, logarithm, or other mathematical function.  Think of them like a machine:



We declare which bits of information are the inputs at in the function definition, inside the parentheses. We can then use the information, called a **parameter** (or many parameters), inside the parentheses for calculations in the function.  For instance, here's a function that squares an input.

```python
def square_this(factor):
    if factor<0:my_square = factor ** 2
    print(my_square)

square_this(3)
```

```python
def square_this(factor):
    if factor<0:
        print("I don't want to square that!")
        
    print(factor ** 2)

square_this(3)
```

In the overly-complicated `square_this` function, there is a single parameter called `factor`.  The body of the function squares whatever value is passed in as `factor`, saves the calculation temporarily in the variable `my_square`, and then prints the results.  

The function call `square_this(3)` calls the function and passes the value 3, known as an **argument**. The argument is passed to the function, which assigns the value of 3 to the variable `factor` before it does anything else. There's like a hidden line of code at the top that makes the assignment `factor = 3`. Then the function block is executed just like any other code, line by line.  You can put for statements, if statement, or whatever content you want inside a function block. If you need to indent code, like for an `if` statement, you'll just indent it an extra level.  

## Exercise 2

Using our geologic time scale exercise from our `if` statement class, create a function that tells you what part of the timescale you're in when you pass the function a numerical age in Ma as an argument.  Your function definition should include a parameter called `age_Ma`.

```python
def geologic_age(age_in_Ma):
    if age_in_Ma < 0:
        print("That's from the future!")
    elif age_in_Ma == 0:
        print("That's literally right now")
    elif age_in_Ma <= 66.02:
        print("That age is in the Cenozoic!")
    elif age_in_Ma <= 252:
        print("That age is in the Mesozoic!")
        print("(Dino kids would love it here!)")
    elif age_in_Ma <= 541:
        print("That age is in the Paleozoic!")
    elif age_in_Ma <= 2500:
        print("That age is in the Proterozoic!")
    elif age_in_Ma <= 4000:
        print("That age is in the Archean!")
    elif age_in_Ma <= 4000:
        print("That age is in the Hadean!")
    elif age_in_Ma >= 4543:
        print("That age is in the protoplanetary disc!")

geologic_age(50000)
```

## Return values

Your function can also return one or more objects (like a number, a list, or something more complicated). This works just like you'd think a math function would work.  When you call the `np.sqrt()` function, you expect Python to return the square root.  You signatl to Python which value to return with the `return` keyword.  The `return` keyword halts all further execution of the function and, if you place an expression afterwards, returns that value.

Returning to my `square_this` function above, if I want to return a the value of the function instead of just printing it to the screen, then I could modify the code like this:

```python
def square_this(factor):
    my_square = factor ** 2
    return my_square

area_square = square_this(3)
extra_area = 5 # hectares

total_area = area_square + extra_area
print(total_area)
```

## Multiple parameters

You can define a function with multiple parameters. Put them all inside the parentheses and separate them with commas.  When you pass arguments to your function, they will be assigned to the parameters in the order you provide them.  You must pass the same numer of arguments as you have parameters!

```python
def quadratic_function(x, a, b, c):
    return a * x**2 + b*x + c

quadratic_function(1, 2, 3, 4)
```
